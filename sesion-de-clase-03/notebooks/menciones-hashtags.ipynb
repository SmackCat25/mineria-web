```python
import json
from collections import Counter
import matplotlib.pyplot as plt
import re
```

```python
with open('../data/posts.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
menciones = []
hashtags = []

for post in data['data']:
    text = post['text']
    # extraer menciones
    menciones.extend(re.findall(r'@(\w+)', text))
    # extraer hashtags
    hashtags.extend(re.findall(r'#(\w+)', text))
```

```python
mention_counts = Counter(menciones)
hashtag_counts = Counter(hashtags)
```

```python
# Mostrar los top 10 menciones y hashtags
top_menciones = mention_counts.most_common(10)
top_hashtags = hashtag_counts.most_common(10)
```

```python
print("Top 10 menciones:")
for mencion, count in top_menciones:
    print(f"{mencion}: {count}")

print("\nTop 10 hashtags:")
for tag, count in top_hashtags:
    print(f"{tag}: {count}")
```

```python
# Menciones
if top_menciones:
    names, counts = zip(*top_menciones)
    plt.figure(figsize=(10, 5))
    plt.bar(names, counts, color='skyblue')
    plt.title("Top 10 Menciones")
    plt.ylabel("Cantidad")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Hashtags
if top_hashtags:
    tags, counts = zip(*top_hashtags)
    plt.figure(figsize=(10, 5))
    plt.bar(tags, counts, color='lightgreen')
    plt.title("Top 10 Hashtags")
    plt.ylabel("Cantidad")
    plt.xticks(rotation=45)
    plt.show()
```
