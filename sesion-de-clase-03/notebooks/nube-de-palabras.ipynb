```python
import json
import re
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
with open('../data/posts2.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
textos = [tweet['text'] for tweet in data['data']]
```

```python
def limpiar_texto(text):
    text = text.lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"@\S+", "", text)
    text = re.sub(r"#\S+", "", text)
    text = re.sub(r"[^\w\s]", "", text)
    return text
```

```python
textos_limpios = [limpiar_texto(t) for t in textos]
todos_textos = " ".join(textos_limpios)
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    max_words=100
).generate(todos_textos)
```

```python
plt.figure(figsize=(15, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
